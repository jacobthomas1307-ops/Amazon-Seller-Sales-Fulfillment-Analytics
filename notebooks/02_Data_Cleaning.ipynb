```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("../data/raw/amazon_sales.csv")
```

```python
df_clean = df.copy()
```

```python
df_clean.head()
```

```python
df_clean.columns
```

```python
df_clean["Date"] = pd.to_datetime(
    df_clean["Date"],
    errors="coerce"
)
```

```python
df_clean["Date"].dtype
```

```python
df_clean["Date"].isnull().sum()
```

```python
df_clean.head()
```

```python
df_clean.info()
```

```python
df_clean["Status"] = df_clean["Status"].str.title()

df_clean["Fulfilment"] = df_clean["Fulfilment"].str.title()

df_clean["Courier Status"] = df_clean["Courier Status"].str.title()
```

```python
missing = (
    df_clean.isnull()
            .sum()
            .sort_values(ascending=False)
)

missing[missing > 0]
```

Drop useless columns

```python
df_clean.drop(
    columns=[
        "fulfilled-by",
        "promotion-ids",
        "Unnamed: 22"
    ],
    inplace=True
)
```

```python
df_clean.info()
```

Handle Amount and Currency

```python
df_clean[
    df_clean["Amount"].isna()
][["Amount","currency"]].head()
```

These rows have:

No sale amount
No currency
Usually correspond to cancelled or invalid transactions
Cannot contribute to revenue calculations

If we keep them:

Revenue calculations become inaccurate.
Average order value gets distorted.
Power BI visuals may show blanks.
SQL aggregations become messier.

Since only about 6% (7,795 rows) are affected, dropping them is a reasonable business decision.

```python
df_clean = df_clean.dropna(subset=["Amount"])
```

```python
df_clean["Amount"].isna().sum()
```

```python
df_clean["currency"].isna().sum()
```

Checking Courier Status

```python
df_clean[
    df_clean["Courier Status"].isna()
][["Status", "Courier Status"]].head(20)
```

Fill it with "Not Shipped"

Missing Courier Status values correspond to cancelled orders that were never shipped. They are replaced with "Not Shipped" to preserve business meaning and simplify downstream analysis.

```python
df_clean["Courier Status"] = df_clean["Courier Status"].fillna("Not Shipped")
```

Handle the remaining 33 missing shipping values

```python
df_clean["ship-country"] = df_clean["ship-country"].fillna("IN")

df_clean["ship-state"] = df_clean["ship-state"].fillna("Unknown")

df_clean["ship-city"] = df_clean["ship-city"].fillna("Unknown")

df_clean["ship-postal-code"] = (
    df_clean["ship-postal-code"]
        .fillna(df_clean["ship-postal-code"].mode()[0])
)
```

```python
missing = (
    df_clean.isnull()
            .sum()
            .sort_values(ascending=False)
)

missing[missing > 0]
```

Standardize Text Values

```python
text_cols = df_clean.select_dtypes(include="object").columns

for col in text_cols:
    df_clean[col] = df_clean[col].str.strip()
```

```python
df_clean["Status"] = df_clean["Status"].str.title()

df_clean["Fulfilment"] = df_clean["Fulfilment"].str.title()

df_clean["Courier Status"] = df_clean["Courier Status"].str.title()
```

Check Duplicate Rows
