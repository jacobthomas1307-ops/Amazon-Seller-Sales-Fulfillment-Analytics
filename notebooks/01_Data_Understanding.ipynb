## 8. Data Dictionary

| Column | Data Type | Description | Business Importance |
|---------|-----------|-------------|---------------------|
| Order ID | Object | Unique identifier for each order | Tracks individual customer orders |
| Date | Date | Date on which the order was placed | Used for time-series and seasonal sales analysis |
| Status | Object | Current order status | Helps analyze delivered, cancelled, and returned orders |
| Fulfilment | Object | Fulfilment method (Amazon or Merchant) | Measures fulfilment performance |
| Sales Channel | Object | Platform through which the order was placed | Evaluates sales source performance |
| ship-service-level | Object | Shipping service level | Measures shipping efficiency |
| Style | Object | Product style/design | Product-level sales analysis |
| SKU | Object | Stock Keeping Unit | Inventory management |
| ASIN | Object | Amazon Standard Identification Number | Unique Amazon product identifier |
| Category | Object | Product category | Identifies top-performing product categories |
| Size | Object | Product size | Inventory and demand planning |
| Courier Status | Object | Shipping progress status | Logistics performance analysis |
| Qty | Integer | Number of items purchased | Sales volume measurement |
| Currency | Object | Currency used in transaction | Financial reporting |
| Amount | Float | Total order value | Revenue analysis |
| ship-city | Object | Customer city | Regional sales analysis |
| ship-state | Object | Customer state | State-wise sales performance |
| ship-postal-code | Float | Customer postal code | Geographic delivery analysis |
| ship-country | Object | Customer country | Country-level sales analysis |
| promotion-ids | Object | Promotion or discount identifier | Marketing campaign analysis |
| B2B | Boolean | Indicates Business-to-Business order | Separates B2B and retail sales |
| fulfilled-by | Object | Fulfilment partner | Supply chain performance |

## Business Observations

- The dataset contains 128,975 Amazon sales records across 24 business attributes.
- Most variables are categorical, reflecting product, customer, and fulfilment information.
- No duplicate records were found, indicating strong data integrity.
- Missing values are primarily concentrated in the fulfilled-by, promotion-ids, Amount, and Currency columns, which will require preprocessing.
- Most customer orders consist of a single item, suggesting predominantly individual purchases.
- Order values show a right-skewed distribution, where a few expensive purchases increase the average order value.
- Geographic information is nearly complete, enabling reliable location-based sales analysis.
- The dataset is suitable for sales performance, fulfilment efficiency, customer behaviour, and regional analytics.
